# Data Visualization with Iris Dataset

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris.species.unique()
```

# KDE PLOT

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),
                     'sepal_length'],color='r',Label='setosa')
sns.set_style('darkgrid')
```

```python
sns.kdeplot(iris.loc[(iris['species']=='setosa'),
                    'sepal_length'],shade=True,color='r',Label='Setosa')
sns.kdeplot(iris.loc[(iris['species']=='versicolor'),
                    'sepal_length'],shade=True,color='b',Label='Versicolor')
sns.set_style('whitegrid')
```

# BOX PLOT

```python
sns.boxplot(x='species',y='sepal_width',data=iris)
```

```python
ax = sns.boxplot(x="species", y="petal_length", data=iris)
ax = sns.stripplot(x="species", y="petal_width", data=iris, jitter=True, edgecolor="gray")
```

```python
ax = sns.boxplot(x="species", y="petal_length", data=iris)
ax = sns.stripplot(x="species", y="petal_length", data=iris, jitter=True, edgecolor="gray")
```

```python
iris.boxplot(by="species", figsize=(12, 6))
pass
```

# BOXEN PLOT

```python
fig=plt.gcf()
fig.set_size_inches(10,7)
fig=sns.boxenplot(x='species',y='sepal_length',data=iris)
```

# FACE GRID

```python
from matplotlib import pyplot as plt
import seaborn as sns
 
iris = sns.load_dataset('iris')
 
sns.set_style("darkgrid")
 
sns.FacetGrid(iris, hue ="species",
              height = 6).map(plt.scatter,
                              'sepal_length',
                              'petal_length').add_legend()
```

# JOINT PLOT

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, size=5)
```

```python
fig=sns.jointplot(x='sepal_length',y='sepal_width',kind='hex',color='r',data=iris)
```

```python
sns.jointplot("petal_length", "petal_width", data=iris, kind="kde",space=0,color='g')
pass
```

# VIOLIN PLOT

```python
sns.violinplot(x="species", y="sepal_length", data=iris, size=6)
```

# COUNT PLOT

```python
import matplotlib.pyplot as plt
sns.countplot('species',data=iris)
plt.show()
```

# PIE CHART

```python
iris['species'].value_counts().plot.pie(explode=[0.1,0.1,0.1],autopct='%1.1f%%',shadow=True,figsize=(10,8))
```

# PAIR PLOT

```python
sns.pairplot(iris,hue='species');
```

# HEAT MAP

```python
plt.figure(figsize=(7,4))
sns.heatmap(iris.corr(),annot=True,cmap='summer')
```

# HISTOGRAM

```python
iris.hist(edgecolor='black',linewidth=.5)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show()
```
